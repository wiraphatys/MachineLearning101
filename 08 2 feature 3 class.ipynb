Workshop: Classification => 2 Features / 2 Classes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/fruit3class.csv')
df.head()
```

```python
sns.scatterplot(data=df, x='m', y='d', hue='target', style='target', s=100, palette='winter')
plt.show()
```

```python
x = df.drop(['target'], axis=1)
x.head()
```

```python
y = df['target']
y.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(x, y)
```

```python
fruitsName = ['Grape','Lemon','Orange']
```

```python
user_input = [[5.6, 4.2],
              [7.6, 4.5],
              [7.8, 5.4]
              ]
```

```python
prediction = model.predict(user_input)
prediction
```

```python
for i, k in enumerate(prediction):
    print(f'{user_input[i]} = {fruitsName[k]}')
```

```python
from sklearn.tree import plot_tree
```

```python
featureName = ['m','d']

plt.subplots(figsize=(10,7), dpi=96)
plot_tree(model, filled=True, rounded=True,
          feature_names=featureName,
          class_names=fruitsName
          )
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
x = np.array(x)
y = np.array(y)
```
