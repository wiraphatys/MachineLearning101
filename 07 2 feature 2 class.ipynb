Workshop: Classification => 2 Features / 2 Classes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from mlxtend.plotting import plot_decision_regions
```

Dataset

```python
df = pd.read_csv('data/fruit2class.csv')
df
```

Data Visualization

```python
sns.scatterplot(data=df, x=df.m, y=df.d, hue='target', style='target', s=100)
sns.set(rc={'figure.figsize':(6,6)})
plt.show()
```

Model

```python
x = df.drop(['target'], axis=1)
x.head()
```

```python
y = df['target']
y.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

Predict

```python
fruitsName = ['grape','lemon']
```

```python
# input = [[m],[d]]
user_input = [[5.6,3.5],[7.6,4.5]]

predictionResult = model.predict(user_input)

for i, key in enumerate(predictionResult):
    print('input: {} => result: {}'.format(user_input[i], fruitsName[key]))
    # print(i, fruitsName[key])
```

Decision Tree Plot

```python
from sklearn.tree import plot_tree
```

```python
data_feature_name = ['m','d']

plt.subplots(nrows=1, ncols=1, figsize=(8, 13), dpi=96)

plot_tree(model,
          filled=True, rounded=True,
          feature_names=data_feature_name,
          class_names=fruitsName
          )
plt.show()
```

Decision regions

```python
from mlxtend.plotting import plot_decision_regions
```

```python
# to array
x = np.array(x)
y = np.array(y)
```

```python
plot_decision_regions(x, y, clf=model)

plt.scatter(user_input[0][0], user_input[0][1], s=150, c='red')
plt.scatter(user_input[1][0], user_input[1][1], s=150, c='black')

plt.xlabel('m')
plt.ylabel('d')
plt.show()
```
