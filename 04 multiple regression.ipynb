Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# check data
df.head()
```

```python
#preparing data
x_axis_data = df[['TV','Radio','Newspaper']]
x_axis_data.head()
```

```python
#preparing data
y_axis_data = df['Sales']
y_axis_data.head()
```

Traing Model & Evaluation

```python
model = LinearRegression()
model.fit(x_axis_data, y_axis_data)
```

```python
model.score(x_axis_data, y_axis_data)
```

```python
y_prediction = model.predict(x_axis_data)
y_prediction
```

```python
r2_score(y_axis_data, y_prediction)
```

```python
mean_absolute_error(y_axis_data, y_prediction)
```

```python
mean_squared_error(y_axis_data, y_prediction)
```
