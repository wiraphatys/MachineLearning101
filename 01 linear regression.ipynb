```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset

```python
x_data = [1.0, 1.8, 3.0, 4.1, 5.2, 6.0]
y_data = [1, 1.3, 2.2, 2.5, 2.8, 3.6]

x = np.array(x_data)
y = np.array(y_data)
```

Data Visualization

```python
sns.scatterplot(x=x_data, y=y_data)
sns.set_style("darkgrid")
plt.title('relational of invested money & return profit')
plt.xlabel('invested money')
plt.ylabel('return profit')
plt.show()
```

```python
x = x.reshape(-1,1)
y = y.reshape(-1,1)
```

Model & Train

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x, y)
```

Predict

```python
new_input = [[2.5],[7]]
```

```python
y_predict = model.predict(new_input)
y_predict
```

```python
model.predict(x)
```

```python
trained_data = model.predict(x)

plt.scatter(x_data, y_data, color="b", label="raw data")
plt.plot(x_data, trained_data, linewidth='1', color='black')
plt.scatter(x_data, trained_data, color='r', marker='s', label="trained data")
plt.legend()
plt.show()
```
