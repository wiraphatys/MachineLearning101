Decision Tree: Classification

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [[27],
     [35],
     [39],
     [21],
     [22],
     [45]
     ]

yName = ['lemon','orange','orange','lemon','lemon','orange']
y = [0, 1, 1, 0, 0, 1]
```

Data Visualization

```python
plt.scatter(yName, x)
plt.xlabel('fruits')
plt.ylabel('size')
plt.show()
```

Model

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

Predict

```python
fruitName = ['lemon','orange']
```

```python
newSize_input = [[25.3]]
```

```python
predictResult = model.predict(newSize_input)
fruitName[predictResult[0]]
```

Visualize Model

```python
from sklearn.tree import plot_tree
```

```python
data_feature_name = ['size']
```

```python
plt.subplots(nrows=1, ncols=1, figsize=(4, 5), dpi=96)

plot_tree(model,
          filled=True, rounded=True,
          feature_names=data_feature_name,
          class_names=fruitName,)
plt.show()
```

ขอบเขตการแบ่ง (Decision regions)

```python
from mlxtend.plotting import plot_decision_regions
```

```python
#plot needs Numpy
x = np.array(x)
y = np.array(y)
```

```python
# plot
plot_decision_regions(x, y, clf=model)
plt.show()
```
